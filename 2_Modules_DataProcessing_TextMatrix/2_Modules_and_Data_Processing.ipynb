\*\* **Important note about references**. In Python, each variable will pass as an object and it depends on the type of object if we can pass it by value or reference (values correspond to "objects" so the variable is just the name). Look at [documentation](http://python-textbok.readthedocs.io/en/1.0/Variables_and_Scope.html) or the following examples to understand this better. Remember that there are objects that we cannot change (their previous data will be collected in "garbage") !!


````python
### Main variable a, reference b
a = ['order1', 'order2']
b = a
print "a =",a," b =", b ,"\n"

### Add something to our main variable and the reference will change
print "Change a main"
a.append("order3")
print "a =",a," b =", b ,"\n"

### Add something to our reference and the main will change
print "Change b reference"
b.append('order4')
print "a =",a," b =", b ,"\n"

### Change the main and then the reference will keep main values
print "Reassign a main"
a = ["NEW ORDER"]
print "a =",a," b =", b 
````

**2.1 Pandas and dataframes** 

Pandas works with the DATAFRAME concept which is one of the main data structures used in data analytics.

A Dataframe is 2-dimensional "labeled" data structure with columns of potentially different types. It is generally the most commonly used pandas object. Along with the data, you can optionally pass index (row labels) and columns (column labels) arguments. More details [here](http://pandas.pydata.org/pandas-docs/stable/dsintro.html#dataframe).

Let's build a dataframe called data_pandas, we'll create some records and put them in the dataframe structure, with the columns labeled:

## 4. Regular expressions and unstructured data 

How can we find all the records that have ID with a "1" ?? 

When trying to work with lines of text, we might be interested in finding patterns to extract information. Searching through files can be done using regular expressions with a tool called grep (Global Regular Expression Printer for the terminal) or regex. In other words, we need to match with ‘regular expressions’ (characters that we want to find in a text), and it's pretty useful for other Natural Language Procesing purposes too.

> In my case, I did this because I have a bunch of transaction logs (records / lines of text coming from a database management system) that I need to classify, and then parse to extract some data. This might be helpful for web crawling, search engines, or any execution list.

> So this is sort of a “query” in a line. For more references you can take a look to [RegEx cheat sheet](https://rexegg.com) and a famous online tool to create the expression for a particular text: [RegExpr](https://regexr.com/).

They define it as  "A regular expression (or RE) specifies a set of strings that matches it; the functions in this module let you check if a particular string matches a given regular expression (or if a given regular expression matches a particular string, which comes down to the same thing)".

> While working with **data, modules and parsing** we can undestand the following example.

> Here are some daily (open source) stock prices with modules such as **urlib2** and **beautifulsoup**. Here, I'll do Amazon's. To do this you should take a look to the website of interest first (you can see the HTML of [historical prices at yahoo finance](https://finance.yahoo.com/quote/amzn/history]) and understand why I am looking for certain patterns.. the word "Date" and "Open" in a "table").

````python

import urllib2
from bs4 import BeautifulSoup as bs

def getYahoo_records(name, numRecords):
    records = []
    url = "https://finance.yahoo.com/quote/" + name + "/history/"
    HTMLtable = bs(  urllib2.urlopen(url).read(), "lxml" ).findAll('table')[0].tbody.findAll('tr')
    for row in HTMLtable:
        if len(records) < numRecords:
            dataPerColumn = row.findAll('td')
            #if dataPerColumn[1].span.text  != 'Dividend': 
            records.append( {'Date': dataPerColumn[0].span.text,
                             'Open': float(dataPerColumn[1].span.text.replace(',',''))} )
    return records

sample_stock = getYahoo_records('amzn', 5)

sample_stock
````

````python

data_stock = pd.DataFrame( getYahoo_records('amzn', 30) )
data_stock['Returns'] =  np.log( data_stock['Open']/data_stock['Open'].shift(1) )
data_stock
````
